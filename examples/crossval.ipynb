```julia
using LowRankModels, DataFrames
import Gadfly

m,n,k = 200,200,3
A = randn(m,k)*randn(k,n) + k*sprandn(m,n,.05)
losses = fill(huber(),n)
r = quadreg(.1)
glrm = GLRM(A,losses,r,r,k+2)

function plot(df, xs::Symbol, ys::Array{Symbol}; scale = :linear, filename=None, height=3, width=6)
    dflong = vcat(map(l->stack(df,l,xs),ys)...)
    if scale ==:log
        p = Gadfly.plot(dflong,x=xs,y=:value,color=:variable,Gadfly.Scale.y_log10)
    else
        p = Gadfly.plot(dflong,x=xs,y=:value,color=:variable)
    end 	
    if !(filename==None)
		println("saving figure in $filename")
        Gadfly.draw(Gadfly.PDF(filename, width*Gadfly.inch, height*Gadfly.inch), p) 
	end
	return p
end
```

```julia
println("Computing regularization path")
train_error, test_error, train_time, model_onenorm, reg_params = 
regularization_path(glrm, params=Params(1,50,.0001,.1), reg_params=linspace(3,0,21), verbose=true)
df = DataFrame(train_error = train_error, test_error = test_error,
               train_time = train_time, model_onenorm = model_onenorm, reg_param = reg_params)
```

```julia
p = plot(df, :reg_param, [:train_error, :test_error], 
scale = :linear, filename="reg_path2_jnb.pdf")
```

```julia
import Gadfly
dflong = vcat(map(l->stack(df,l,:reg_param), [:train_error, :test_error])...)
p = Gadfly.plot(dflong,x=:reg_param,y=:value,color=:variable)
```

```julia
draw(PDF("regpath.pdf", 6inch, 3inch), p) 
```
